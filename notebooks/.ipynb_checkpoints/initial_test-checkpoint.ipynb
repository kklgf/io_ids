# Imports and setup

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
%config Completer.use_jedi = False
```

# Parameters

```python
path_train = "../data/kddcup.data"
path_test = "../data/corrected"
path_names = "../data/kddcup.names"
# path_types = "../data/training_attack_types.txt"
```

# Load labels and datasets

```python
# types_df = pd.read_csv(path_types, header=None, sep=" ")
# types_types = list(types_df[0])
# types_types.append("normal")
# print(f"len(types_types): {len(types_types)}")
```

```python
names_types = []
names_names = []
with open(path_names) as f:
    for i, line in enumerate(f.readlines()):
        if i == 0:
            names_types = line.replace('.','').strip().split(',')
        else:
            names_names.append(line.replace(':','').strip().split(' ')[0])
names_names.append("label")
print(f"len(names_types): {len(names_types)}, len(names_names): {len(names_names)}")
```

```python
train_df = pd.read_csv(path_train, header=None, names=names_names)
train_df.head()
```

```python
test_df = pd.read_csv(path_test, header=None, names=names_names)
test_df.head()
```

# Look into data

```python
non_numerical_names = ['protocol_type', 'service', 'flag', 'label']
```

```python
train_df.describe()
```

```python
test_df.describe()
```

```python
for name in non_numerical_names:
    unique = train_df[name].unique()
    unique.sort()
    print(f"====================================================================")
    print(f"================== {'train_df':8}:{name:>16}:{len(unique):>4} ==================")
    for i in range(0, len(unique), 4):
        print(''.join(f"{e:>16}|" for e in unique[i:i+4]))
    unique = test_df[name].unique()
    unique.sort()
    print(f"================== {'test_df':8}:{name:>16}:{len(unique):>4} ==================")
    for i in range(0, len(unique), 4):
        print(''.join(f"{e:>16}|" for e in unique[i:i+4]))
    print("\n\n")
```
